```python
# importing required libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
# reading a csv file with ara and its prices for training a model
df = pd.read_csv("D:\\Group_Projects\\linear_regression\\area.csv")
```

```python
df.head()
```

```python
# splitting dependent and independent features as X and y. It should be of dataframe 
X = df[['area']]
y = df[['price']]

print(X)
print(y)
type(X)
```

```python
# plotting X and y to understand the data better
plt.plot(X, y, color = 'b', marker = 'o')
plt.xlabel('Area in Sqft')
plt.ylabel('Price in USD')
plt.title('SFO House Prices')
plt.grid()
plt.show()
```

```python
# training a linear regression model on X and y features 
model = linear_model.LinearRegression()
model.fit(X,y)
```

```python
# model coefficients or parameters
model.coef_
```

```python
model.predict([[4500]])
```

```python
# reading a file with area for predictions
df1 = pd.read_csv('predict_prices.csv')
```

```python
#predicted house prices for the given area and added a predicted price column back to the same file
df1['predicted_prices'] = model.predict(df1[['area']])
```

```python
model.predict([[4000]])
```

```python
# comparing the predicted values with corresponding to the original values
plt.plot(X, y, color = 'b', marker = 'x')
plt.plot(df1.area, df1.predicted_prices, color = 'r', marker = 'o')
plt.xlabel('Area in Sqft')
plt.ylabel('Price in USD')
plt.title('SFO House Prices')
plt.grid()
plt.show()
```

```python
# saving the file with areas and predicted houses as a csv file
df1.to_csv('predicted_house_prices')
```
